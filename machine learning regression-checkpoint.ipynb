# gradient descent

```python
import numpy as np
def gradient(x,y):
    m_curr=b_curr=0
    iterations=500
    n=len(x)
    learning_rate=0.05
    
    for i in range(iterations):
        y_pred=m_curr*x+b_curr
        J=(1/2*n)*sum([val**2 for val in (y-y_pred)])
        md=-(2/n)*sum(x*(y-y_pred))
        bd=-(2/n)*sum(y-y_pred)
        m_curr=m_curr-learning_rate*md
        b_curr=b_curr-learning_rate*bd
        print(m_curr," ",b_curr," ",i," ",J)
x=np.array([1,2,3,4,5])
y=np.array([5,7,9,11,13])
gradient(x,y)
```
